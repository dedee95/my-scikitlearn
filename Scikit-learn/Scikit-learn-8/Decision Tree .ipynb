# Decision Tree Classification
Decision tree merupakan suatu metode klasifikasi yang menggunakan struktur pohon, dimana setiap node merepresentasikan atribut dan cabangnya merepresentasikan nilai dari atribut, sedangkan daunnya digunakan untuk merepresentasikan kelas. Node teratas dari decision tree ini disebut dengan root. <br>
<strong>root node </strong>= fear terminology <br>
<strong> internal node </strong> = sejumlah node yang berada di antara root node dan leaf node  <br>
<strong> leaf node</strong>  = hasil keputusan/ prediksi paling akhir sudah tidak ada cabang (daun) <br>

### Persiapan Dataset

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print(f'Dimensi feature: {X.shape}')
print(f'Dimensi target: {y.shape}')
```

```python
# spliting dataset ke train dan test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=.3,
                                                    random_state=0)

print(f'Dimensi X_train: {X_train.shape}')
print(f'Dimensi X_test: {X_test.shape}')
print(f'Dimensi y_train: {y_train.shape}')
print(f'Dimensi y_test: {y_train.shape}')
```

### Training Model

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4) # decisien memiliki kedalaman maksimum 4 layer

model.fit(X_train, y_train)
```

### Visualisasi Model

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.rcParams['figure.dpi'] = 85
plt.subplots(figsize=(10, 10))
tree.plot_tree(model, fontsize=10)

plt.show()
```

##### Catatan
didalam value terdapat 3 nilai artinya kelas indeks ke 0 terdapat x data, kelas indeks ke 1 terdapat y data, dan kelas indeks ke 2 terdapat z data. 

### Model Evaluation

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
